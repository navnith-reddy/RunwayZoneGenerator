```python
import runwayZoneGenerator as rzg
import pandas as pd

# Initialise runway centreline dataset
rzg.buildRunways()
runways = rzg.readRunways()

# Sydney Dataset
sydney = runways.loc[runways['NAME'] == 'SYDNEY AIRPORT']
sydney.reset_index(drop=True, inplace=True)

# Canberra Dataset
canberra = runways.loc[runways['NAME'] == 'CANBERRA AIRPORT']
canberra.reset_index(drop=True, inplace=True)

# Canberra Dataset
melbourne = runways.loc[runways['NAME'] == 'MELBOURNE AIRPORT']
melbourne.reset_index(drop=True, inplace=True)

# Canberra Dataset
perth = runways.loc[runways['NAME'] == 'PERTH INTERNATIONAL AIRPORT']
perth.reset_index(drop=True, inplace=True)
```

```python
# Initialise empty list of naming variables
name = []

# Initialise lists for other data
location = ['Sydney'] * 12
identifier = ['YSSY'] * 12

# Retrieve runway centreline geometry from dataset
rwcl16l = sydney.at[0, 'geometry']
# Create zone geometries
rw16l = rzg.asymmetricZones(rwcl16l, side=1, opFreq='3950')
# Create naming variables
name.extend(['16L']*3)

# Repeat steps above, but with welded runways
# Step not required for datasets without bisected runways
rwcl16r = rzg.weldRunways(sydney.at[1, 'geometry'], sydney.at[2, 'geometry'])
rw16r = rzg.asymmetricZones(rwcl16r, side=1, opFreq='3950')
name.extend(['16R']*3)

rw34r = rzg.asymmetricZones(rwcl16l, side=0, opFreq='3950')
name.extend(['34R']*3)

rw34l = rzg.asymmetricZones(rwcl16r, side=0, opFreq='3950')
name.extend(['34L']*3)

# Merge runways and add columns
sydRunways = rw16l.append(rw16r, ignore_index=True)
sydRunways = sydRunways.append(rw34r, ignore_index=True)
sydRunways = sydRunways.append(rw34l, ignore_index=True)
sydRunways['Location'] = location
sydRunways['Identifier'] = identifier
sydRunways['Runway'] = name
sydRunways.explore()
```

```python
name = []
location = ['Perth'] * 6
identifier = ['YPPH'] * 6

rwcl03 = rzg.weldRunways(perth.at[2, 'geometry'], perth.at[4, 'geometry'])
rw03 = rzg.asymmetricZones(rwcl03, side=1, opFreq='3950')
name.extend(['03']*3)

rw21 = rzg.asymmetricZones(rwcl03, side=0, opFreq='3950')
name.extend(['21']*3)

# Merge runways and add columns
perthRunways = rw03.append(rw21, ignore_index=True)
perthRunways['Location'] = location
perthRunways['Identifier'] = identifier
perthRunways['Runway'] = name
perthRunways.explore()
```

```python
name = []
location = ['Canberra'] * 3
identifier = ['YSCB'] * 3

rwcl35 = rzg.weldRunways(canberra.at[0, 'geometry'], canberra.at[3, 'geometry'])
rw35 = rzg.zoneConstructor(rwcl35, opFreq='3950')
name.extend(['35']*3)

cbrRunways = rw35
cbrRunways['Location'] = location
cbrRunways['Identifier'] = identifier
cbrRunways['Runway'] = name
cbrRunways.explore()
```

```python
name = []
location = ['Melbourne'] * 3
identifier = ['YMML'] * 3

rwcl16 = rzg.weldRunways(melbourne.at[0, 'geometry'], melbourne.at[3, 'geometry'])
rw16 = rzg.zoneConstructor(rwcl16, opFreq='3950')
name.extend(['16']*3)

melbRunways = rw16
melbRunways['Location'] = location
melbRunways['Identifier'] = identifier
melbRunways['Runway'] = name
melbRunways.explore()
```

```python
runways = sydRunways.append(cbrRunways, ignore_index=True)
runways = runways.append(melbRunways, ignore_index=True)
runways = runways.append(perthRunways, ignore_index=True)
columnNames = ['Location', 'Identifier', 'Runway', 'Polygon Name', 'geometry']
runways = runways.reindex(columns=columnNames)
runways.to_file('study/study.shp')
runways
```

```python
exclusionZones = runways.loc[runways['Polygon Name'] == 'Exclusion Zone']
exclusionZones.reset_index(drop=True, inplace=True)
exclusionZones.to_file('study/exclusionZones.shp')

restrictionZones = runways.loc[runways['Polygon Name'] == 'Restriction Zone']
restrictionZones.reset_index(drop=True, inplace=True)
restrictionZones.to_file('study/restrictionZones.shp')

runwayCentrelines = runways.loc[runways['Polygon Name'] == 'Runway']
runwayCentrelines.reset_index(drop=True, inplace=True)
runwayCentrelines.to_file('study/runwayCentrelines.shp')
```
